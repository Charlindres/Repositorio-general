```python
import os
 
dirname = "C:\\Windows"
mitexto = os.listdir(dirname)
 
f = open("fichero.txt", 'w')
for elemento in mitexto:
    print elemento
    f.write(elemento)
    f.write("\r\n")
f.close()
```
